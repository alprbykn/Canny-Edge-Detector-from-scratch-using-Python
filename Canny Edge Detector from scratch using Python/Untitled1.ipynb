```python
from scipy import ndimage
import skimage.feature
import skimage.io as skio
import imageio
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showIm(img):
    plt.imshow(img,cmap=plt.get_cmap('gray'))
    return plt.show()
```

```python
def gaussian_Filter(img):
    return ndimage.gaussian_filter(img,sigma=1.5)
```

```python
def cannys(img):
    showIm(img)
    img=gaussian_Filter(img)
    showIm(img)
    return 0


img = imageio.imread("C:/Users/user/Desktop/dersler/imageProcessing/assignment_2/Assignment 2/chihuahua-dog-puppy-cute-39317_small.jpeg")
img=img.astype('int32')
cannys(img)
```
